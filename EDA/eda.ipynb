```python
# 1. Chargement des données traitées
import pandas as pd
df = pd.read_csv('../data/processed/Japan/Tokyo.csv')
df.head()
```

```python
"""
2. Analyse des distributions
    Exemples:
        - Histogramme des température
        - Évolution du score météo par mois
"""

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Température (°C)"], kde=True, color='skyblue')
plt.title("Distribution des températures à Tokyo")
plt.xlabel("Température (°C)")
plt.ylabel("Fréquence")
plt.grid(True)
plt.tight_layout() # Ajuste automatiquement l'espacement
plt.show()
```

```python
"""
    3. Corrélation météo
        Tu peux montrer s'il y a des liens entre:
            - Tempéraure <-> score météo
            - Humidité <-> score météo
"""
sns.heatmap(df[["Température (°C)", "Humidité (%)", "Vent (m/s)", "Score Météo"]], annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
import calendar

top_months = df.sort_values(by="Score Météo", ascending=False).reset_index(drop=True)
top_months["Nom du mois"] = top_months["Mois"].apply(lambda x: calendar.month_name[x])
top_months["Rang"] = top_months.index + 1 # Crée une colonne rang
print("Top 10 mois recommandés pour voyager à Tokyo :")
print(top_months[["Rang", "Année", "Nom du mois", "Score Météo"]].head(10)) # top 10
```
